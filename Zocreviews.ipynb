```python
import pandas as pd
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

#zocreviews_df = pd.read_csv('~/Documents/Bootcamp/zocdoc_reviews/zocdocreviews/zocdocreviews.csv')
zocreviews_df = pd.read_csv('~/Documents/Bootcamp/zocdoc_reviews/zocdocreviews/zocdocreviews2-151-182.csv')
```

```python
#zocreviews2_df = pd.read_csv('~/Documents/Bootcamp/zocdoc_reviews/zocdocreviews/zocdocreviews182_214.csv')
zocreviews2_df = pd.read_csv('~/Documents/Bootcamp/zocdoc_reviews/zocdocreviews/zocdocreviews2-182-220.csv')
zocreviews2_df
```

```python
#zocreviews3_df=pd.read_csv('~/Documents/Bootcamp/zocdoc_reviews/zocdocreviews/zocdocreviews214_251.csv')
zocreviews3_df=pd.read_csv('~/Documents/Bootcamp/zocdoc_reviews/zocdocreviews/zocdocreviews2-220-251.csv')
```

```python
zocrev_df=pd.concat([zocreviews_df, zocreviews2_df, zocreviews3_df], axis = 0)
    
```

```python
zocrev_df=zocrev_df.drop_duplicates()
zocrev_df=zocrev_df.dropna()
zocrev_df
```

```python
#zocdocs_df = pd.read_csv('~/Documents/Bootcamp/zocdoc/zocdocdoctors_clean.csv')
```

```python
zocdocs_df2=zocdocs_df.dropna(subset=['doctor_type','education','board_certs','gender','individual_rating','npi','overall_patient_ratings','overall_rating','practice','street_address','city_state','zoc_awards'], how='all')
zocdocs_df2=zocdocs_df2.drop_duplicates()
zocdocs_df2['overall_patient_ratings'] = zocdocs_df2['overall_patient_ratings'].str.strip('[]')
zoc_df3=zocdocs_df2.join(zocdocs_df2['overall_patient_ratings'].str.split(',', 2, expand=True).rename(columns={0:'overall_p_rating', 1:'wait_rating', 2:'bedside_rating'}))
zoc_df3
```

```python
zoc_df3.drop('overall_p_rating', axis=1, inplace=True)
zoc_df3.drop('gender', axis=1, inplace=True)
zoc_df3.drop('individual_rating', axis=1, inplace=True)
zoc_df3.drop('npi', axis=1, inplace=True)
zoc_df3.drop('overall_patient_ratings', axis=1, inplace=True)
zoc_df3.drop('street_address', axis=1, inplace=True)
zoc_df3.drop('city_state', axis=1, inplace=True)
#dropping unnecessary columns
```

```python
zoc_df3['wait_rating']=pd.to_numeric(zoc_df3.wait_rating)
zoc_df3['bedside_rating']=pd.to_numeric(zoc_df3.bedside_rating)
zoc_df3['overall_rating']=pd.to_numeric(zoc_df3.overall_rating)
zoc_df3
```

```python
mergedzoc_df=pd.merge(zocreviews2_df, zoc_df3, how='left', on ='doctor').dropna()
mergedzoc_df
```

```python
five_rating=mergedzoc_df.loc[(mergedzoc_df['overall_rating'] ==5.0)]
four_rating=mergedzoc_df.loc[(mergedzoc_df['overall_rating'] >= 4.0) & (mergedzoc_df['overall_rating'] < 5.0)]
three_rating=mergedzoc_df.loc[(mergedzoc_df['overall_rating'] >= 3.0) & (mergedzoc_df['overall_rating'] < 4.0)]
two_rating=mergedzoc_df.loc[(mergedzoc_df['overall_rating'] >= 2.0) & (mergedzoc_df['overall_rating'] < 3.0)]
```

```python
five_rating['text']=five_rating['text'].str.replace('doctor',"")
five_rating['text']=five_rating['text'].str.replace('Doctor',"")
five_rating['text']=five_rating['text'].str.replace('Sequeira',"")
five_rating['text']=five_rating['text'].str.replace('Dr',"")
four_rating['text']=four_rating['text'].str.replace('Doctor',"")
four_rating['text']=four_rating['text'].str.replace('doctor',"")
four_rating['text']=four_rating['text'].str.replace('Dr',"")

#df['range'] = df['range'].str.replace(',','-')
five_rating
```

```python
fivewait_rating=mergedzoc_df.loc[(mergedzoc_df['wait_rating'] ==5.0)]
fourwait_rating=mergedzoc_df.loc[(mergedzoc_df['wait_rating'] >= 4.0) & (mergedzoc_df['wait_rating'] < 5.0)]
threewait_rating=mergedzoc_df.loc[(mergedzoc_df['wait_rating'] >= 3.0) & (mergedzoc_df['wait_rating'] < 4.0)]
threewait_rating['text']=threewait_rating['text'].str.replace('Dr',"")
threewait_rating['text']=threewait_rating['text'].str.replace('Doctor',"")
threewait_rating['text']=threewait_rating['text'].str.replace('doctor',"")
```

```python
fivebedside_rating=mergedzoc_df.loc[(mergedzoc_df['bedside_rating'] ==5.0)]
fourbedside_rating=mergedzoc_df.loc[(mergedzoc_df['bedside_rating'] >= 4.0) & (mergedzoc_df['bedside_rating'] < 5.0)]
threebedside_rating=mergedzoc_df.loc[(mergedzoc_df['bedside_rating'] >= 3.0) & (mergedzoc_df['bedside_rating'] < 4.0)]
```

```python
fivebedside_rating['text']=fivebedside_rating['text'].str.replace('doctor',"")
fivebedside_rating['text']=fivebedside_rating['text'].str.replace('Doctor',"")
fivebedside_rating['text']=fivebedside_rating['text'].str.replace('Dr',"")
threebedside_rating['text']=threebedside_rating['text'].str.replace('Dr',"")
threebedside_rating['text']=threebedside_rating['text'].str.replace('Doctor',"")
threebedside_rating['text']=threebedside_rating['text'].str.replace('doctor',"")
threebedside_rating['text']=threebedside_rating['text'].str.replace('Chynn',"")
```

```python


from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS

#mpl.rcParams['figure.figsize']=(8.0,6.0)    #(6.0,4.0)
mpl.rcParams['font.size']=12                #10 
mpl.rcParams['savefig.dpi']=100             #72 
mpl.rcParams['figure.subplot.bottom']=.1 


stopwords = set(STOPWORDS)
data = four_rating

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords.add("Icasiano"),
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(data['text']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```
